```python
import pandas as pd

companies = pd.read_csv("seeds/companies.csv")
emissions = pd.read_csv("seeds/emissions.csv")
initiatives = pd.read_csv("seeds/initiatives.csv")
target_trajectory = pd.read_csv("seeds/target_trajectory.csv")
```

```python
print("=== COMPANIES ===")
print(companies.info())
print("\n")
print(companies.head())
print("\n")
print("Duplicates:", companies.duplicated().sum())
print("\n")
print("Nulls:\n", companies.isnull().sum())
```

```python
print("=== EMISSIONS ===")
print(emissions.info())
print("\n")
print(emissions.head())
print("\n")
print("Duplicates:", emissions.duplicated().sum())
print("\n")
print("Nulls:\n", emissions.isnull().sum())
print("\n")
print("Stats:\n", emissions.describe())
```

```python
print("=== DATA TYPES EMISSIONS ===")
print(emissions.dtypes)
```

```python
print("=== EMISSIONS NULLS ===")
print(emissions[emissions['emissions_tco2e'].isna()])
```

```python
print("=== WHOLE LINE EMISSIONS DUPLICATES ===")
print(emissions[emissions.duplicated(subset=['company_id', 'year', 'scope', 'category'], keep=False)])
```

```python
print("=== INVALID YEARS ===")
print(emissions[emissions['year'] > 2026])
```

```python
print("=== NEGATIVE EMISSIONS ===")
print(emissions[emissions['emissions_tco2e'] < 0])
```

```python
print("=== INVALID SCOPE ===")
print(emissions[emissions['scope'] == 4])
```

```python
print("=== INVALID UNITS ===")
print(emissions['unit'].value_counts())
print("\n")
print(emissions[emissions['unit'] == 'kgCO2e'])
```

```python
print("=== TARGETS ===")
print(target_trajectory.info())
print("\n")
print(target_trajectory.head())
print("\n")
print("Duplicates:", target_trajectory.duplicated().sum())
print("\n")
print("Nulls:\n", target_trajectory.isnull().sum())
print("\n")
print("Stats:\n", target_trajectory.describe())
```

```python
print("=== SBTI VALIDATED REDUCTIONS ===")
print(target_trajectory[
    (target_trajectory['sbti_validated'] == True) & 
    (target_trajectory['reduction_pct'] < 0.42) &
    (target_trajectory['target_year'] == 2030)
][['target_id', 'company_id', 'reduction_pct', 'target_year', 'sbti_validated']].drop_duplicates())
```

```python
print("=== TARGET TRAJECTORY CHARACTERISTICS ===")
print("scope_coverage:", target_trajectory['scope_coverage'].value_counts())
print("\n")
print("method:", target_trajectory['method'].value_counts())
print("\n")
print("unit:", target_trajectory['unit'].value_counts())
```

```python
first_year = target_trajectory.groupby('target_id').agg(
    first_year=('year', 'min'),
    baseline_year=('baseline_year', 'first')
).reset_index()

print("=== FIRST YEAR != BASELINE YEAR ===")
print(first_year[first_year['first_year'] != first_year['baseline_year']])
```

```python
last_year = target_trajectory.groupby('target_id').agg(
    last_year=('year', 'max'),
    target_year=('target_year', 'first')
).reset_index()

print("=== LAST YEAR != TARGET YEAR ===")
print(last_year[last_year['last_year'] != last_year['target_year']])
```

```python
def check_gaps(group):
    years = sorted(group['year'].tolist())
    expected = list(range(min(years), max(years) + 1))
    missing = set(expected) - set(years)
    return missing

gaps = target_trajectory.groupby('target_id').apply(check_gaps)
print("=== YEAR GAPS ===")
for target_id, missing in gaps.items():
    if missing:
        print(f"{target_id} : missing years → {missing}")
```

```python
print("=== INITIATIVES ===")
print(initiatives.info())
print("\n")
print(initiatives.head())
print("\n")
print("Duplicates:", initiatives.duplicated().sum())
print("\n")
print("Nulls:\n", initiatives.isnull().sum())
print("\n")
print("Stats:\n", initiatives.describe())
```

```python
print("=== WHOLE LINE INITIATIVES DUPLICATES ===")
print(initiatives[initiatives.duplicated(keep=False)])
```

```python
print("=== INITIATIVE GROUP TYPE ===")
print(initiatives['group_type'].value_counts())
print("\n")
print("=== INITIATIVE GROUP TYPE THAT HAVE NOT NULL PARENT GROUP ID===")
print(initiatives.groupby('group_type')['parent_group_id'].count())
print("\n")
print("=== BOTTOM-UP INITIATIVES (PARENT AND CHILDREN) ===")
print(initiatives[initiatives['group_type'].isin(['bottom_up_group', 'bottom_up_child'])]
    [['initiative_id', 'group_type', 'parent_group_id', 'estimated_reduction']])
```

```python
print("=== INITIATIVE STATUS ===")
print(initiatives['status'].value_counts())
print("\n")
print("=== INITIATIVE LEVER CATEGORY ===")
print(initiatives['lever_category'].value_counts())
print("\n")
print("=== INITIATIVE UNITS ===")
print(initiatives['unit'].value_counts())
```

```python
print("=== HIGH ESTIMATED REDUCTION ===")
print(initiatives[initiatives['estimated_reduction'] > 1000000])
```

```python
print("=== INITIATIVE INVALID COMPANY_ID ===")
print(initiatives[~initiatives['company_id'].isin(companies['company_id'])])
print("\n=== INITIATIVE INVALID TARGET_ID ===")
print(initiatives[~initiatives['target_id'].isin(target_trajectory['target_id'])])
```

```python
print("=== INITIATIVE START YEAR > END YEAR ===")
print(initiatives[initiatives['start_year'] > initiatives['end_year']])
```

```python
print("=== DUPLICATES INITIATIVE_ID GRAIN ===")
print(initiatives[initiatives.duplicated(subset=['initiative_id'], keep=False)])
print("\nNombre:", initiatives.duplicated(subset=['initiative_id']).sum())

print("\n=== DUPLICATES COMPANIES GRAIN ===")
print(companies.duplicated(subset=['company_id']).sum())

print("\n=== DUPLICATES EMISSIONS GRAIN ===")
print(emissions.duplicated(subset=['company_id', 'year', 'scope', 'category']).sum())

print("\n=== DUPLICATES TARGETS GRAIN ===")
print(target_trajectory.duplicated(subset=['target_id', 'year']).sum())
```
